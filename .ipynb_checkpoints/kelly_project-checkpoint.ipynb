```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import datetime
from scipy import stats
import os

print(os.listdir("Resources/names"))
```

```python
year = []
for filename in os.listdir("Resources/names"):
    if 'yob20' in filename:
        namefile = pd.read_csv(f"Resources/names/{filename}",header=None)
        namefile['year'] = os.path.splitext(filename)[0][-4:]
        year.append(namefile)

#data is in series, merge all years together
df = pd.concat(year)
```

```python
df.head()
```

```python
name_df = df.rename(columns={
    0:'name',
    1:'gender',
    2:'count'
})
name_df.head()
```

```python
name_df.dtypes
```

```python
#count of total babies born in the year
yearTotal_df = name_df.groupby('year').sum().rename(columns={'count':'yearTotal'})
```

```python
yearTotal_df.head()
```

```python
#merge total babies born in year and unique names
name_df = name_df.merge(yearTotal_df,on='year')
name_df.head()
```

```python
#Calculate % of name
name_df['name_percent'] = 100*(name_df.loc[:,'count']/name_df.loc[:,'yearTotal'])
name_df.head()
```

```python
#Get top 5 female names in year 2011
f2011_df1 = name_df[name_df['gender']=='F'] 
f2011_df2 = f2011_df1[f2011_df1['year']=='2011'].sort_values('name_percent',ascending=False)
top5_f2011_df = f2011_df2.head()
top5_f2011_df.head()
```

```python
top5_f2011_df["Rank"] = top5_f2011_df["name_percent"].rank(ascending=False) 
top5_f2011_df.head()
```

```python
#Trace the same name in 2017
f2017_df1 = name_df[(name_df['gender']=='F') & (name_df['year']=='2017') & 
                   ((name_df['name']=='Sophia') |
                    (name_df['name']=='Isabella') |
                    (name_df['name']=='Emma') |
                    (name_df['name']=='Olivia') |
                    (name_df['name']=='Ava')
                   )
                   ]

f2017_df1
```

```python
f2017_df1["Rank"] = f2017_df1["name_percent"].rank(ascending=False) 
f2017_df1.head()
```

```python
#union - top 5 is same for 2011 and 2017
#result = pd.concat([top5_f2011_df,f2017_df1])
#result
```

```python
#merge 2011 vs 2017
merge = top5_f2011_df.merge(f2017_df1,on='name')
merge['rank_change'] = merge['Rank_x'] - merge['Rank_y']
merge
```

```python
#change of rank from 2011 vs 2017
plt.bar(merge['name'],merge['rank_change'])
plt.ylabel('Change in Rank from 2011 to 2017')
plt.title('Top 5 Female Names in 2011 vs 2017')
```
